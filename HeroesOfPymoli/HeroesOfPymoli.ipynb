```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
purchase_data_df = "Resources/purchase_data.csv"
```

```python
purchase_data_df = pd.read_csv(purchase_data_df) # Read Purchasing File and store into Pandas data frame
purchase_data_df #display purchase data dataframe
```

```python
#Display the total number of players
total_players_df=pd.DataFrame({
    "Total Players":[purchase_data_df['SN'].nunique()]
})
total_players_df 
```

```python
#Run basic calculations to obtain number of unique items, average price, etc.
unique_items = len(purchase_data_df["Item ID"].unique())
average_price = purchase_data_df["Price"].mean()
Number_purchases = purchase_data_df ["Purchase ID"].count()
Total_revenue = purchase_data_df ["Price"].sum()

#Create a summary data frame to hold the results
summary_data_df = pd.DataFrame({
    "Number of Unique Items": [unique_items],
    "Average Price": [average_price],
    "Number of Purchases": [Number_purchases],
    "Total Revenue": [Total_revenue]
})

###############################################
#Optional: give the displayed data cleaner formatting

summary_data_df["Average Price"] = summary_data_df["Average Price"].map("${:,.2f}".format)
summary_data_df["Total Revenue"] = summary_data_df["Total Revenue"].map("${:,.2f}".format)

#Display the summary data frame

summary_data_df[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]
```

```python
#Gender Demographics
#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

#Gender percentage
gender_df = purchase_data_df.groupby(["Gender"])
gender_count = gender_df["SN"].nunique()
gender_percentage = gender_count / total_players_df ['Total Players'][0]*100

#Gender summary table 
gender_temp_df = pd.DataFrame({"Total Count": gender_count,
                               "Percentage of Players": gender_percentage
}) 

#Data formatting
gender_temp_df ["Percentage of Players"] = gender_temp_df["Percentage of Players"].map("{:.2f}%".format)

#Summary data frame
gender_temp_df[["Total Count", "Percentage of Players"]].sort_values(["Total Count"], ascending=False)
```

```python
#Purchasing Analysis (Gender)
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame


Purchase_count = gender_df["Purchase ID"].count()
Average_purchase_price = gender_df["Price"].mean()
Total_purchase_value = gender_df["Price"].sum()
Average_total_purchase = Total_purchase_value / gender_count



Gender_purchase_df = pd.DataFrame({"Purchase Count": Purchase_count,
                                "Average Purchase Price": Average_purchase_price,
                                "Total Purchase Value": Total_purchase_value,
                                "Avg Total Purchase per Person": Average_total_purchase
})
                                   
Gender_purchase_df["Average Purchase Price"] = Gender_purchase_df["Average Purchase Price"].map("${:,.2f}".format)
Gender_purchase_df["Total Purchase Value"] = Gender_purchase_df["Total Purchase Value"].map("${:,.2f}".format)
Gender_purchase_df["Avg Total Purchase per Person"] = Gender_purchase_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

Gender_purchase_df[["Purchase Count",
                 "Average Purchase Price",
                 "Total Purchase Value",
                 "Avg Total Purchase per Person"]].sort_values(["Purchase Count"], ascending=False)
```

```python
#Age Demographics
#Establish bins for ages
#Categorize the existing players using the age bins. Hint: use pd.cut()
#Calculate the numbers and percentages by age group
#Create a summary data frame to hold the results
#Optional: round the percentage column to two decimal points
#Display Age Demographics Table

bins = [0, 9, 14, 19, 24, 29, 34, 39, 150] #bin range
age_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]#bin labels


purchase_data_df["Age Labels"] = pd.cut(purchase_data_df["Age"], bins, labels=age_labels)
age_group_df = purchase_data_df.groupby(["Age Labels"])

age_count = age_group_df["SN"].nunique()
age_percentage = age_count / total_players_df ['Total Players'][0]* 100

age_temp_df =pd.DataFrame({"Total Count": age_count,
                            "Percentage of Players": age_percentage
})

age_temp_df["Percentage of Players"] = age_temp_df["Percentage of Players"].map("{:.2f}%".format)
age_temp_df[["Total Count", "Percentage of Players"]]
```

```python
#Purchasing Analysis (Age)
#Bin the purchase_data data frame by age
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
#Create a summary data frame to hold the results
#Optional: give the displayed data cleaner formatting
#Display the summary data frame

bins=[0,9,14,19,24,29,34,39,150]
age_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
categorize_age=pd.cut(purchase_data_df['Age'], bins, labels=age_labels, include_lowest=True)

purchase_data_df['Categorized Age']=categorize_age

ages_group=purchase_data_df.groupby('Categorized Age')
purchase_age_count=ages_group['Purchase ID'].count()

avg_purchase_price_age=ages_group['Price'].mean()
total_purchase_age_value=ages_group['Price'].sum()
players_age=ages_group['SN'].nunique()

average_purchase_age= ages_group['Price'].sum()/players_age


age_purchase_df = pd.DataFrame({"Purchase Count": purchase_age_count,
                                "Average Purchase Price": avg_purchase_price_age,
                                "Total Purchase Value": total_purchase_age_value,
                                "Avg Total Purchase per Person": average_purchase_age
})

age_purchase_df["Average Purchase Price"] = age_purchase_df["Average Purchase Price"].map("${:,.2f}".format)
age_purchase_df["Total Purchase Value"] = age_purchase_df["Total Purchase Value"].map("${:,.2f}".format)
age_purchase_df["Avg Total Purchase per Person"] = age_purchase_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

age_purchase_df.rename_axis(None,inplace=True)
age_purchase_df
```

```python
#Top Spenders
#Run basic calculations to obtain the results in the table below
#Create a summary data frame to hold the results
#Sort the total purchase value column in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the summary data frame

grouped_spenders_df = purchase_data_df.groupby(["SN"])

purchase_count_sp = grouped_spenders_df["Purchase ID"].count()
average_purchase_price = grouped_spenders_df["Price"].mean() 
total_purchase_value = grouped_spenders_df["Price"].sum()

spenders_df = pd.DataFrame({"Purchase Count": purchase_count_sp,
                            "Average Purchase Price": average_purchase_price,
                            "Total Purchase Value": total_purchase_value
})

spenders_df = spenders_df.sort_values(["Total Purchase Value"], ascending=False)

spenders_df["Average Purchase Price"] = spenders_df["Average Purchase Price"].map("${:,.2f}".format) 
spenders_df["Total Purchase Value"] = spenders_df["Total Purchase Value"].map("${:,.2f}".format)

spenders_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]].head()
```

```python
#Retrieve the Item ID, Item Name, and Item Price columns
#Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value
#Create a summary data frame to hold the results
#Sort the purchase count column in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the summary data frame

items_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
grouped_items_df = items_df.groupby(["Item ID", "Item Name"])

purchase_count_sp = grouped_items_df["Item ID"].count()
item_price = grouped_items_df["Price"].mean() 
total_purchase_value = grouped_items_df["Price"].sum()

items_df = pd.DataFrame({"Purchase Count": purchase_count_sp,
                         "Item Price": item_price,
                         "Total Purchase Value": total_purchase_value})

items_df = items_df.sort_values(["Purchase Count"], ascending=False)

items_df["Item Price"] = items_df["Item Price"].map("${:,.2f}".format) 
items_df["Total Purchase Value"] = items_df["Total Purchase Value"].map("${:,.2f}".format)

items_df[["Purchase Count", "Item Price", "Total Purchase Value"]].head()
```

```python
#Most Profitable Items
#Sort the above table by total purchase value in descending order
#Optional: give the displayed data cleaner formatting
#Display a preview of the data frame

items_df["Total Purchase Value"] = items_df["Total Purchase Value"].apply(lambda x: x.replace('$', '').replace(',', '')).astype('float')
items_df = items_df.sort_values(["Total Purchase Value"], ascending=False)

items_df["Total Purchase Value"] = items_df["Total Purchase Value"].map("${:,.2f}".format)
items_df[["Purchase Count", "Item Price", "Total Purchase Value"]].head()
```
